```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {
    'Vendor': ['A','B','C','A','B','C','A','B','C','A','B','C'],
    'Month': ['Jan','Jan','Jan','Feb','Feb','Feb','Mar','Mar','Mar','Apr','Apr','Apr'],
    'Delivery_Days': [5,7,6,4,8,5,6,7,5,4,9,6],
    'Cost': [1000,1200,1100,1050,1250,1150,1020,1300,1120,1080,1350,1180],
    'Quantity': [50,60,55,52,65,58,54,62,57,53,68,59]
}

df = pd.DataFrame(data)

print("Dataset Preview:")
display(df)

print("\nChecking for missing values:")
print(df.isnull().sum())

total_cost = df['Cost'].sum()
print("\nTotal Procurement Cost:", total_cost)

# Average Delivery Time
avg_delivery = df['Delivery_Days'].mean()
print("Average Delivery Days:", round(avg_delivery,2))

# Vendor-wise Cost
vendor_cost = df.groupby('Vendor')['Cost'].sum()
print("\nVendor-wise Cost:\n", vendor_cost)

# Vendor Order Count
order_count = df['Vendor'].value_counts()


# 1. Vendor Cost Analysis (Bar Chart)
plt.figure()
vendor_cost.plot(kind='bar')
plt.title("Vendor-wise Total Cost")
plt.xlabel("Vendor")
plt.ylabel("Cost")
plt.show()

# 2. Monthly Cost Trend (Line Chart)
monthly_cost = df.groupby('Month')['Cost'].sum()
plt.figure()
monthly_cost.plot(marker='o')
plt.title("Monthly Procurement Cost Trend")
plt.xlabel("Month")
plt.ylabel("Cost")
plt.show()

# 3. Delivery Performance (Bar Chart)
delivery_avg = df.groupby('Vendor')['Delivery_Days'].mean()
plt.figure()
delivery_avg.plot(kind='bar')
plt.title("Average Delivery Time per Vendor")
plt.xlabel("Vendor")
plt.ylabel("Days")
plt.show()

# 4. Cost Distribution (Pie Chart)
plt.figure()
vendor_cost.plot(kind='pie', autopct='%1.1f%%')
plt.title("Cost Distribution by Vendor")
plt.ylabel("")
plt.show()


print("\n--- INSIGHTS ---")

best_vendor = delivery_avg.idxmin()
print("Best Vendor (Fastest Delivery):", best_vendor)

highest_cost_vendor = vendor_cost.idxmax()
print("Highest Spending Vendor:", highest_cost_vendor)

print("Recommendation: Focus on vendors with lower delivery time and cost efficiency.")
```
